```python
import numpy as np
import matplotlib.pyplot as plt
from liteczos import twosites
from liteczos.lanczos import get_green_function

t = 1
U = 3

omega = np.linspace(-8, 8, 1000)
eta = 0.1

H = twosites.hamiltonian(t, U)
green = get_green_function(H)
expected_green = twosites.get_green_function(t, U)

mu = U/2

fig, ax = plt.subplots(2)
ax[0].set_title("Real")
ax[0].plot(omega, green(omega+1j*eta - mu).real, label="lanczos")
ax[0].plot(omega, expected_green(omega+1j*eta).real, label="analytical")
ax[1].plot(omega, -green(omega+1j*eta - mu).imag, label="lanczos")
ax[1].plot(omega, -expected_green(omega+1j*eta).imag, label="analytical")
ax[1].set_title("Imag")
ax[1].legend()

plt.tight_layout()
plt.show()
```
