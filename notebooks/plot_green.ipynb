```python
import numpy as np
import matplotlib.pyplot as plt
from liteczos import twosites, lanczos

# Parameters
t = 1
U = 4
mu = U/2
eta = 0.1
omega = np.linspace(0, 8, 1000)

# Green functions
expected_green = twosites.get_green_function(t, U)
H = twosites.hamiltonian(t, U)
green = lanczos.get_green_function(H)

# Figures
fig, ax = plt.subplots(2)

ax[0].set_title("Real")
ax[0].plot(omega, green(omega+1j*eta - mu).real, lw=2, label="lanczos")
ax[0].plot(omega, expected_green(omega+1j*eta).real, label="analytical")
ax[0].legend()

ax[1].set_title("Imag")
ax[1].plot(omega, -green(omega+1j*eta - mu).imag, lw=2, label="lanczos")
ax[1].plot(omega, -expected_green(omega+1j*eta).imag, label="analytical")
ax[1].legend()

plt.tight_layout()
plt.show()
```
